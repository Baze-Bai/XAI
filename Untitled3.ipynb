## Import packages:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.inspection import PartialDependenceDisplay
from alibi.explainers import ALE
from sklearn.datasets import fetch_california_housing
```

## Load California Housing dataset

```python
california = fetch_california_housing()
df = pd.DataFrame(california.data, columns=california.feature_names)
df["MedHouseVal"] = california.target
```

## Exploratory analysis

```python
# Calculate correlation coefficient
correlation_matrix = df.corr()

# Heat map of correlation
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5) ## assisted by GPT-4o at 23:09pm 2/5 2025
plt.title("Correlation Matrix of California House Price")
plt.show()

```

The correlation matrix shows relationships between features in the California Housing Dataset. Key insights include: AveRooms has the strongest positive correlation (0.85) with AveBedrms, indicating that higher income areas tend to have higher house prices. Average Rooms (AveRooms) has a weaker positive correlation (0.15) with house prices, while Latitude and Longitude show strongest negative correlation, demostrating their opposite relationship. MedInc and MedHouseVal also have a strong correlation with 0.85, which is resonable.

## Develop Random Forest Regression Model

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["MedHouseVal"]), df["MedHouseVal"], test_size=0.2, random_state=42)

# Train the Random Forest Model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Calculate the feature importance
feature_importances = pd.Series(model.feature_importances_, index=X_train.columns).sort_values(ascending=False)

# Demonstrate the importance of the feature
plt.figure(figsize=(8, 5))
sns.barplot(x=feature_importances.values, y=feature_importances.index, palette="viridis")
plt.title("Feature Importance of Random Forest")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.show()
```

We can see the top 2 important features are MedInc and AveOccup.

## Plot PDP and ICE

```python
# Select the top2 important features for PDP and ICE
top_features = feature_importances.index[:2]
# Plot PDP and ICE
fig, ax = plt.subplots(figsize=(12, 6))
PartialDependenceDisplay.from_estimator(model, X_test, features=top_features, kind="both", ax=ax) ## assisted by GPT-4o at 23:15pm 2/5 2025
plt.suptitle("PDP and ICE", fontsize=14)
plt.show()
```

This PDP and ICE plot visualizes the impact of Median Income (MedInc) and Average Occupancy (AveOccup) on house prices. The left plot shows a strong positive relationship between MedInc and house prices, indicating that higher income leads to increased house prices. The right plot shows a negative relationship between AveOccup and house prices, suggesting that higher average occupancy per household correlates with lower house prices. The ICE lines reveal individual variations, with some non-linearity in both features.

## Plot ALCE

```python
# Generare ALE plot
ale_explainer = ALE(model.predict, feature_names=list(X_train.columns))
ale_exp = ale_explainer.explain(X_test.values)

# Plot ALE
fig, ax = plt.subplots(ncols=2, figsize=(14, 5))
for i, feature in enumerate(top_features): ## assisted by GPT-4o at 23:20pm 2/5 2025
    ax[i].plot(ale_exp.feature_values[i], ale_exp.ale_values[i])
    ax[i].set_title(f"Figure ALE: {feature}")
    ax[i].set_xlabel(feature)
    ax[i].set_ylabel("ALE influence")
plt.tight_layout()
plt.show()
```

This PDP and ICE plot visualizes the impact of Median Income (MedInc) and Average Occupancy (AveOccup) on house prices. The left plot shows a strong positive relationship between MedInc and house prices, indicating that higher income leads to increased house prices. The right plot shows a negative relationship between AveOccup and house prices, suggesting that higher average occupancy per household correlates with lower house prices. The ICE lines reveal individual variations, with some non-linearity in both features.
